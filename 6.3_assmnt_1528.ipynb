#Task Description #2: Loops (Multiples of a Number)
Scenario
You are writing a utility function to display multiples of a given number.

```python
def print_multiples_for(number):
    """
    Prints the first 10 multiples of the input number using a for loop.
    - range(1, 11) generates numbers 1 through 10.
    - Multiply each by the input number and print.
    """
    print(f"First 10 multiples of {number} (for loop):")
    for i in range(1, 11):  # Loop runs 10 times exactly
        print(number * i)

# Example usage
print_multiples_for(3)
```

```python
def print_multiples_while(number):
    """
    Prints the first 10 multiples using a while loop.
    - Initialize counter i=1.
    - Loop until i > 10, printing and incrementing each time.
    """
    print(f"First 10 multiples of {number} (while loop):")
    i = 1  # Start counter
    while i <= 10:  # Condition checks each iteration
        print(number * i)
        i += 1  # Increment to avoid infinite loop

# Example usage
print_multiples_while(3)
```

#Task Description #4: For and While Loops (Sum of First n Numbers)
Scenario
You need to calculate the sum of the first n natural numbers.

```python
def sum_to_n_for_loop(n):
    """
    Calculates the sum of the first n natural numbers using a for loop.
    """
    if n < 0:
        return "Input must be a non-negative integer"
    total_sum = 0
    for i in range(1, n + 1):
        total_sum += i
    return total_sum

# Test cases
print(f"Sum of first 5 natural numbers: {sum_to_n_for_loop(5)}")
print(f"Sum of first 10 natural numbers: {sum_to_n_for_loop(10)}")
print(f"Sum of first 0 natural numbers: {sum_to_n_for_loop(0)}")
print(f"Sum of first -3 natural numbers: {sum_to_n_for_loop(-3)}")
```

```python
def sum_to_n_while_loop(n):
    """
    Calculates the sum of the first n natural numbers using a while loop.
    """
    if n < 0:
        return "Input must be a non-negative integer"
    total_sum = 0
    count = 1
    while count <= n:
        total_sum += count
        count += 1
    return total_sum

# Test cases
print(f"Sum of first 5 natural numbers (while loop): {sum_to_n_while_loop(5)}")
print(f"Sum of first 10 natural numbers (while loop): {sum_to_n_while_loop(10)}")
print(f"Sum of first 0 natural numbers (while loop): {sum_to_n_while_loop(0)}")
print(f"Sum of first -3 natural numbers (while loop): {sum_to_n_while_loop(-3)}")
```

Both the while loop and the for loop are used to perform repetitive tasks, but they differ in how they control the iteration process. In a while loop, the initialization and increment must be done manually, and the loop continues to execute as long as the given condition remains true. This gives the while loop more flexibility, as it can be used for situations where the number of repetitions is not known in advance. However, it also increases the chance of errors, such as infinite loops, if the condition or increment is not handled properly.

On the other hand, the for loop is best suited for count-controlled iterations where the number of repetitions is known. It automatically handles initialization, condition checking, and increment through the use of the range() function, making the code shorter and easier to read. In the case of printing the first ten multiples of a number, the for loop is more efficient and readable, while the while loop offers a more general-purpose structure. Both approaches produce the same output, but the for loop is preferred for fixed iteration tasks due to its simplicity and reduced chances of logical errors.

2.  **Initialization:** Two variables are initialized: `total_sum` to 0 (to store the cumulative sum) and `count` to 1 (as the starting number for summation).
3.  **Condition Check:** The `while` loop continues as long as `count` is less than or equal to `n`.
4.  **Accumulation and Increment:** Inside the loop, the current `count` is added to `total_sum`, and then `count` is incremented by 1. This ensures that the loop progresses through each number from 1 to `n`.
5.  **Return Value:** Once the `while` loop condition becomes false (i.e., `count` is greater than `n`), the loop terminates, and `total_sum` is returned.

#Task Description #3: Conditional Statements (Age Classification)
Scenario
You are building a basic classification system based on age.

```python
def classify_age_nested_if_elif_else(age):
    """
    Classifies an individual's age into categories using nested if-elif-else statements.
    """
    if age < 0:
        return "Invalid Age"
    else:
        if age <= 12:
            return "Child"
        elif age <= 17:
            return "Teenager"
        elif age <= 64:
            return "Adult"
        else:
            return "Senior"

# Test cases
print(f"Age 5: {classify_age_nested_if_elif_else(5)}")
print(f"Age 15: {classify_age_nested_if_elif_else(15)}")
print(f"Age 30: {classify_age_nested_if_elif_else(30)}")
print(f"Age 70: {classify_age_nested_if_elif_else(70)}")
print(f"Age -1: {classify_age_nested_if_elif_else(-1)}")
```

### Explanation of Nested if-elif-else Logic:

This function `classify_age_nested_if_elif_else` takes an `age` as input and returns a classification string. The logic proceeds as follows:

1.  **Initial Check (Invalid Age):** It first checks if `age` is less than 0. If true, it immediately returns "Invalid Age" because age cannot be negative.

2.  **Nested Classification:** If the age is not invalid (i.e., `age >= 0`), it enters a nested `if-elif-else` block to categorize the age:
    *   **Child:** `if age <= 12`: If the age is 12 or less, it's classified as a "Child".
    *   **Teenager:** `elif age <= 17`: If the age is greater than 12 (due to the previous condition) but 17 or less, it's a "Teenager".
    *   **Adult:** `elif age <= 64`: If the age is greater than 17 but 64 or less, it's an "Adult".
    *   **Senior:** `else`: If none of the above conditions are met (meaning age is greater than 64), it's classified as a "Senior".

This structure ensures that each age falls into exactly one category, with a specific order of checks.

### Alternative: Simplified if-elif-else for Age Classification

This approach uses a single `if-elif-else` chain, which can often be more readable for sequential condition checks where each `elif` implicitly relies on the previous `if` or `elif` being false. This eliminates the need for nesting by ordering the conditions from most restrictive to least restrictive, or vice versa, based on the logic.

```python
def classify_age_simplified_if_elif_else(age):
    """
    Classifies an individual's age into categories using a simplified if-elif-else structure.
    """
    if age < 0:
        return "Invalid Age"
    elif age <= 12:
        return "Child"
    elif age <= 17:
        return "Teenager"
    elif age <= 64:
        return "Adult"
    else:
        return "Senior"

# Test cases
print(f"Age 5: {classify_age_simplified_if_elif_else(5)}")
print(f"Age 15: {classify_age_simplified_if_elif_else(15)}")
print(f"Age 30: {classify_age_simplified_if_elif_else(30)}")
print(f"Age 70: {classify_age_simplified_if_elif_else(70)}")
print(f"Age -1: {classify_age_simplified_if_elif_else(-1)}")
```

### Comparison of Conditional Structures (Nested vs. Simplified):

Both `classify_age_nested_if_elif_else` and `classify_age_simplified_if_elif_else` achieve the same outcome, but they differ in their structure and readability.

*   **Nested `if-elif-else`:** The first function uses an initial `if-else` block to handle the `Invalid Age` case, and then places a separate `if-elif-else` structure *inside* the `else` block. This nesting can be useful for grouping related conditions or when the initial check completely separates a large branch of logic. However, for a series of mutually exclusive conditions like age ranges, it can sometimes make the code appear slightly more complex.

*   **Simplified `if-elif-else`:** The second function uses a single, flat `if-elif-else` chain. The conditions are evaluated sequentially. Once a condition is met, the corresponding block is executed, and the rest of the `elif` and `else` blocks are skipped. This structure is often preferred for its linear flow and better readability when handling a series of distinct, ordered conditions. The `elif age <= 12` implicitly means `age >= 0` (because `age < 0` was caught by the initial `if`), `elif age <= 17` implicitly means `age > 12`, and so on.

In this specific age classification scenario, the **simplified `if-elif-else`** chain is generally considered more Pythonic and easier to read because the conditions are naturally ordered, and the lack of nesting reduces visual complexity.

```python
class BankAccount:
    """
    A simple BankAccount class to demonstrate object-oriented programming concepts.
    It allows users to deposit, withdraw, and check their balance.
    """

    def __init__(self, account_holder, initial_balance=0):
        """
        Initializes a new BankAccount instance.

        Args:
            account_holder (str): The name of the account holder.
            initial_balance (float, optional): The initial balance of the account.
                                             Defaults to 0. Cannot be negative.
        """
        if initial_balance < 0:
            raise ValueError("Initial balance cannot be negative.")
        self.account_holder = account_holder
        self.balance = initial_balance
        print(f"Account for {self.account_holder} created with initial balance: ${self.balance:.2f}")

    def deposit(self, amount):
        """
        Deposits a specified amount into the account.

        Args:
            amount (float): The amount to be deposited. Must be positive.
        """
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.balance += amount
        print(f"Deposited: ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        """
        Withdraws a specified amount from the account.

        Args:
            amount (float): The amount to be withdrawn. Must be positive and not exceed current balance.
        """
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if amount > self.balance:
            print(f"Insufficient funds. Current balance: ${self.balance:.2f}. Attempted withdrawal: ${amount:.2f}")
            return
        self.balance -= amount
        print(f"Withdrew: ${amount:.2f}. New balance: ${self.balance:.2f}")

    def check_balance(self):
        """
        Displays the current balance of the account.
        """
        print(f"Current balance for {self.account_holder}: ${self.balance:.2f}")

# --- Demonstration of the BankAccount Class ---

# Create a new account
my_account = BankAccount("Alice Smith", 1000)

# Check initial balance
my_account.check_balance()

# Perform some deposits
my_account.deposit(500)
my_account.deposit(200.75)

# Check balance after deposits
my_account.check_balance()

# Perform some withdrawals
my_account.withdraw(300)
my_account.withdraw(1500) # Attempt to overdraw
my_account.withdraw(50) # Another valid withdrawal

# Check final balance
my_account.check_balance()

# Attempt to create an account with a negative initial balance
try:
    invalid_account = BankAccount("Bob", -50)
except ValueError as e:
    print(f"Error creating account: {e}")
```

This `BankAccount` class is designed to simulate basic banking operations. Here's a breakdown of its components and how it works:

1.  **`__init__(self, account_holder, initial_balance=0)` (Constructor):**
    *   This is a special method called when a new `BankAccount` object is created.
    *   It takes `account_holder` (the name of the account owner) and an optional `initial_balance` (defaulting to 0) as arguments.
    *   **Input Validation:** It checks if the `initial_balance` is negative. If so, it raises a `ValueError` to prevent an invalid account state.
    *   It initializes two instance attributes: `self.account_holder` and `self.balance`.
    *   A message is printed confirming the account creation.

2.  **`deposit(self, amount)` Method:**
    *   This method allows adding funds to the account.
    *   **Input Validation:** It checks if the `amount` to deposit is positive. If not, it prints an error message.
    *   If the amount is valid, it adds the `amount` to `self.balance`.
    *   It then prints the deposited amount and the new balance.

3.  **`withdraw(self, amount)` Method:**
    *   This method allows removing funds from the account.
    *   **Input Validation:** It performs two checks:
        *   Ensures the `amount` to withdraw is positive.
        *   Checks if the `amount` is less than or equal to the `self.balance` to prevent overdrawing.
    *   If both conditions are met, it subtracts the `amount` from `self.balance`.

4.  **`check_balance(self)` Method:**
